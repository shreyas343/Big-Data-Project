```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
Customers = pd.read_csv("Customer_data.csv")
Sales = pd.read_csv("Sales_data.csv")
Product = pd.read_csv("Product_data.csv")
```

```python
Customers
```

```python
Customers.drop_duplicates(inplace=True)
```

```python
Customers
```

```python
Cust = Customers.dropna()
Cust
```

```python
Sales.drop_duplicates(inplace=True)
Sales
```

```python
Txn = Sales.dropna()
Txn
```

```python
Product.drop_duplicates(inplace=True)
Product
```

```python
Prod = Product.dropna()
Prod
```

```python
Data= pd.merge(left = Txn, right = Prod, on = ["Product_category_code","Product_subcategory_code"], how = "left")
Data
```

```python
Data.isnull().sum()
```

```python
Final = pd.merge(left = Data, right = Customers, on = ["Customer_id"], how = "left")
Final
```

```python
Final.isnull().sum()
```

```python
Final["Customer_dob"] = pd.to_datetime(Final["Customer_dob"],format = "%d-%m-%Y")
Final["Date_of_transaction"] = pd.to_datetime(Final["Date_of_transaction"],format = "%d-%m-%Y")
```

```python
Popular_products = Final.groupby(["Customer_gender","Product_category"])[["Quantity"]].sum().reset_index()
Popular_products.pivot(index = "Customer_gender",columns = "Product_category",values = "Quantity")
```

```python
Cust_Grp = Final.groupby(["Customer_cityCode"])["Customer_id"].count().sort_values(ascending = False)
Cust_Grp
```

```python
plt.figure(figsize = (10,7))
plt.title("Maximum number of customers in each city")
Cust_Grp.plot(kind = "bar")
plt.xlabel("City")
plt.ylabel("No. of Customers")
plt.yticks(np.arange(0 , 2500, step = 200))
plt.show()
```

```python
# a = Final.groupby(["Store_type"])["Quantity","Price"].sum().sort_values(by="Quantity",ascending = False)
a = Final.groupby(["Store_type"])[["Quantity", "Price"]].sum().sort_values(by="Quantity", ascending=False)
a
```

```python
sns.set_style("dark")
a.plot.bar(rot=0, subplots=True, color={"Quantity": "green", "Price": "purple"})
plt.legend()
plt.show()
```

```python
# a.plot.pie(subplots=True, title = 'Products sold - Store type (Quantity and Price)', colors = ['pink', 'mediumaquamarine', 'darkgrey', 'skyblue'], figsize=(20, 10))

import matplotlib.pyplot as plt

# Create a figure and axes with specific spacing between subplots
fig, axes = plt.subplots(1, len(a.columns), figsize=(20, 10), subplot_kw=dict(aspect="equal"))

# Plot the pie charts
a.plot.pie(
    subplots=True, 
    ax=axes,  # Assign the axes array to handle multiple charts
    colors=['pink', 'mediumaquamarine', 'darkgrey', 'skyblue'], 
    autopct='%1.1f%%',  # Optional to show percentages
    labeldistance=1.2,  # Adjust label distance to avoid overlapping
    textprops={'fontsize': 18}  # Increase label font size
)

# Set the figure's main title
fig.suptitle('Products sold - Store type (Quantity and Price)', fontsize=20)

# Loop through axes and adjust their labels and titles
for ax, col in zip(axes, a.columns):
    ax.set_ylabel('', fontsize=15)  # Remove ylabel since it's not necessary
    ax.set_title(col, fontsize=15)  # Set title for each chart

# Adjust the layout for better separation
plt.subplots_adjust(wspace=0.4)  # Increase the horizontal space between pie charts

# Tight layout to prevent overlap and add space for the title
plt.tight_layout(rect=[0, 0, 1, 0.95])

plt.show()

```

```python
import matplotlib.pyplot as plt

# Create a figure with a larger size
plt.figure(figsize=(30, 15), dpi=80)  # Increase height to match width

# Adjust subplot spacing
plt.subplots_adjust(hspace=0.5, wspace=0.3)  # Adjust spacing between subplots

# Define larger font sizes
axis_font_size = 24  # Larger font size for axis labels
legend_font_size = 20  # Larger font size for legends
tick_font_size = 18  # Larger font size for tick labels

# Create subplots in a 3x2 grid layout
ax1 = plt.subplot(3, 2, 1)
ax1.plot(q1['Bags'], color="blue", label='Bags')
ax1.legend(loc='upper right', fontsize=legend_font_size)  # Adjust legend size
ax1.tick_params(axis='both', which='major', labelsize=tick_font_size)  # Adjust tick label size

ax2 = plt.subplot(3, 2, 2)
ax2.plot(q1['Books'], color="green", label='Books')
ax2.legend(loc='upper right', fontsize=legend_font_size)  # Adjust legend size
ax2.tick_params(axis='both', which='major', labelsize=tick_font_size)  # Adjust tick label size

ax3 = plt.subplot(3, 2, 3)
ax3.plot(q1['Clothing'], color="red", label='Clothing')
ax3.legend(loc='upper right', fontsize=legend_font_size)  # Adjust legend size
ax3.tick_params(axis='both', which='major', labelsize=tick_font_size)  # Adjust tick label size

ax4 = plt.subplot(3, 2, 4)
ax4.plot(q1['Electronics'], color="pink", label='Electronics')
ax4.legend(loc='upper right', fontsize=legend_font_size)  # Adjust legend size
ax4.tick_params(axis='both', which='major', labelsize=tick_font_size)  # Adjust tick label size

ax5 = plt.subplot(3, 2, 5)
ax5.plot(q1['Footwear'], color="orange", label='Footwear')
ax5.legend(loc='upper right', fontsize=legend_font_size)  # Adjust legend size
ax5.tick_params(axis='both', which='major', labelsize=tick_font_size)  # Adjust tick label size

ax6 = plt.subplot(3, 2, 6)
ax6.plot(q1['Home and kitchen'], color="skyblue", label='Home and Kitchen')
ax6.legend(loc='upper right', fontsize=legend_font_size)  # Adjust legend size
ax6.tick_params(axis='both', which='major', labelsize=tick_font_size)  # Adjust tick label size

# Adjust layout to prevent overlapping
plt.tight_layout()

# Show the plot
plt.show()
```
